```python
import pandas as pd


df = pd.read_csv('/content/pm25_data.csv')

df.head()
```

```python
print(df.columns)
```

```python
# meltを使用してlong formatに変換
df_long = df.reset_index().melt(id_vars=["timestamp"], var_name="location", value_name="value")
df_long = df_long[df_long['location'] != 'index']
df_long
```

```python
df_long.shape
```

```python
#Replace the NaN values ​​to zero.
df = df_long
df.fillna(0, inplace=True)

# Replace the values ​​greater than or equal to 100 to zero
# (only in numeric columns)
for col in df.select_dtypes(include=['number']):
  df.loc[df[col] >= 100, col] = 0

#Remove the "timestamp" column.
#df.drop('timestamp', axis=1, inplace=True)

# カラム名の変更: 'timestamp' を 'tid' に
cleaned_df = df.rename(columns={'timestamp': 'tid'})
cleaned_df = cleaned_df.rename(columns={'location': 'item'})

cleaned_df
```

```python
# Count the number of zero values
num_zeros = (df == 0).sum().sum()

print(f"The number of zero values in the DataFrame is: {num_zeros}")
```

```python

# Importing the required libraries
from PAMI.frequentPattern.basic import FPGrowth as alg
import pandas as pd

# Specify the input parameters
inputFile = '/content/PM24HeavyPollutionRecordingSensors.csv'
separator = '\t'
minimumSupportCountList = [100, 150, 200, 250, 300]

# Initializing an empty DataFrame to store the results
result = pd.DataFrame(columns=['algorithm', 'minSup', 'patterns', 'runtime', 'memory'])

# Iterating over each minimum support count
for minSupCount in minimumSupportCountList:
    # Initializing the FPGrowth algorithm
    obj = alg.FPGrowth(inputFile, minSup=minSupCount, sep=separator)
    # Starting the mining process
    obj.mine()
    # Storing the results in the DataFrame
    result = pd.concat([result, pd.DataFrame({'algorithm': 'FPGrowth', 'minSup': minSupCount,
                                              'patterns': len(obj.getPatterns()), 'runtime': obj.getRuntime(),
                                              'memory': obj.getMemoryRSS()}, index=[0])], ignore_index=True)

# Printing the result
print(result)
```
